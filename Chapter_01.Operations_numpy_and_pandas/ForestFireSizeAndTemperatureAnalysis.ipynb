```python
import pandas as pd
```

```python
dataset = pd.read_csv('../Datasets/forestfires.csv')
dataset.head(2)
```

### Derive insights from the sizes and forest fires

```python
datasetFiltered = dataset[dataset['area'] > 0]
print(dataset.shape)
print(datasetFiltered.shape)
```

```python
mean = datasetFiltered['area'].mean()
print(mean)
```

```python
std = datasetFiltered['area'].std()
print(std)
```

```python
minValue = datasetFiltered.sort_values(by=['area'])['area'].iloc[0]
maxValue = datasetFiltered.sort_values(by=['area'])['area'].iloc[-1]
print(minValue)
print(maxValue)
```

```python
datasetFiltered.describe()
```

```python
datasetSorted = datasetFiltered.sort_values(by=['area'])
datasetSorted.tail(20)
```

```python
datasetSorted['area'].median()
```

### Filtering the month with the most forest fires

```python
months = dataset["month"].unique()
months
```

```python
dataset[dataset["month"] == "mar"].shape[0]
```

```python
for month in months:
    month_dataset = dataset[dataset["month"] == month]
    fires_in_month = month_dataset.shape[0]
    avg_tmp_in_month = int(month_dataset["temp"].mean())
    
    print(str(fires_in_month) + " fires in " + month + " with a mean temperature of ~" + str(avg_tmp_in_month) + "°C")
```
