```python
file = '../Datasets/world_population.csv'
# index_col=0, especifica que queremos cargar la primer columna como indice en lugar de uno autogeneradp 0...n
dataset = pd.read_csv(file, index_col=0)
```

### Using pandas to compute mean, median and variance

```python
# First two rows 
dataset[0:2]
```

```python
# index the third row and using axis compute the mean in the row
# iloc allow us to use the index as an integer
dataset.iloc[[2]].mean(axis=1)
```

```python
# Compute the mean for the last row
dataset.iloc[[-1]].mean(axis=1)
```

```python
# With loc we can index a row with the label
dataset.loc[["Germany"]].mean(axis=1)
```

```python
# compute the median for the last row
dataset.iloc[[-1]].median(axis=1)
```

```python
# Compute the median for the last thre countries in the dataset
dataset[-3:].median(axis=1)
```

```python
# Compute the median for the 10 rows 
dataset.head(10).median(axis=1)
```

```python
# it has to be the same as
dataset[:10].median(axis=1)
```

```python
# Compute the variance of the dataset and return only the last 5 columns
dataset.var().tail()
```

```python
# Compute the variance of the dataset but for rows instead of columns and return only the last 5
dataset.var(axis=1).tail()
```

```python
# NumPy pandas interoperability
import numpy as np
print("pandas", dataset["2015"].mean())
print("numpy", np.mean(dataset["2015"]))
```

### Filtering, sorting and reshaping
#### Filtering

```python
# filtering columns 1961, 2000, and 2015
dataset.filter(items=["1961", "2000", "2015"]).head()
```

```python
# filtering countries that had a greater population density than 500 in 2000
dataset[(dataset["2000"] > 500)][["2000"]]
```

```python
#Get columns that start with 2
dataset.filter(regex="^2", axis=1).head()
```

```python
# Filter rows that start with A
dataset.filter(regex="^A", axis=0).head()
```

```python
#Filter the countries that has land in their names
dataset.filter(like="land", axis=0).head()
```

#### Sorting

```python
# Sort ascending the values in 1961, return only the column 1961 
dataset.sort_values(by=["1961"])[["1961"]].head(10)
```

```python
# Sort ascending the values in 2015, return only the column 2015 
dataset.sort_values(by=["2015"])[["2015"]].head(10)
```

```python
# Sort descending the values in 2015, return only the column 2015 
dataset.sort_values(by=["2015"], ascending=False)[["2015"]].head(10)
```

#### Reshaping

```python
# reshaping to 2015 as row and country codes as columns
dataset_2015 = dataset[["Country Code", "2015"]]
dataset_2015.pivot(index=["2015"] * len(dataset_2015),
                   columns="Country Code",
                   values="2015")
```
